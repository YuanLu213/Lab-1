```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
from scipy import signal

plt.rcParams["figure.figsize"] = (15,10)

print("Probability, Distributions, and Statistical Questions")
print(" ")
print("Problem 1")
print("A. Show how the probability distribution of the background changes as you integrate (sum) for more")
print(" ")
print("Let's assume that the average cosmic-ray background is X = 1 and the average number of gamma-rays emitted by hypothetical source is Y = 5, and the background distribution is Poisson Distribution.")
print("Let us plot the probability mass function of the cosmic-ray background for the first three days. We will be using convolution operator to convolve the distribution with itself for 3 times.")

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharey = True)
plt.xlim(0,10)
mu = 1
x = np.arange(10)
pmf = stats.poisson.pmf(x, mu)
ax1.bar(x,pmf)
day2 = signal.convolve(pmf, pmf)
x = np.arange(len(day2))
ax2.bar(x,day2)
day3 = signal.convolve(day2, day2)
x = np.arange(len(day3))
ax3.bar(x, day3)
plt.title("Probability Mass Function for cosmic background")
plt.xlabel("Observed cosmic rays")
plt.ylabel("Probability")
plt.show()
```

```python
print("The above graphs showed that after integrating for three days, the probability distribution of background is a poisson distribution as it is for day 1.")
print(" ")
print("B. Show that after 5 days, the summed probability distribution is still a Poisson distribution. Explain why this makes sense from a mathematical and conceptual point of view.")
print(" ")
print("To show that, after 5 days, the summed probability distribution is still a Poisson distribution, we will keep repeating the method in part A for two more days.")
fig, ax = plt.subplots(1, 1)
day4 = signal.convolve(day3, day3)
day5 = signal.convolve(day4, day4)
x = np.arange(len(day5))
plt.title("Probability Mass Function for cosmic background summed for 5 days")
plt.xlabel("Observed cosmic rays")
plt.ylabel("Probability")
ax.bar(x, day5)
plt.show()
```

```python
print("As shown above, after 5 days, the summed distribution is still a poisson distribution. This makes sense because, conceptually, the convolution is just to calculate each discrete probability over a interval, and it will not affect or change the functions themselve. Mathematically speaking, convolution is a mathematical operation on two functions that generate a third function to understand how the two functions modified each other. In our case, we are convolving two function both with poisson distributioin, so the third function will also be a poisson distribution.")
print(" ")
print("C. Show how the probability distribution evolves as you average days. Calculate for many different ranges of days, and explore the shape of the distribution as the number of days becomes larger. Discuss this in relation to both B) and the central limit theoem.")
print(" ")
```

```python
print("To do this, we still use convolution operator, and we will use for loop to convolve the distribution 10 times with itself.")
fig, ax = plt.subplots(1,1)
mu = 1
x = np.arange(10)
pmf = stats.poisson.pmf(x, mu)
days = pmf
for x in range (10):
    days = signal.convolve(days, pmf)
    x = np.arange(len(days))
    ax.bar(x, days)
    
plt.title("Probability Mass Function for cosmic background averaged for 10 days")
plt.xlabel("Observed cosmic rays")
plt.ylabel("Probability")
plt.show()
```

```python
print(" ")
print("Problem 2")
print("A. Show how the distribution changes as you average over more observing intervals.")
print(" ")
print("Let us assume that we have a Rayleigh distribution for our cosmic-ray background. To see how the probability density function changes as being averaged for some interval, we first plot the probability density function, and then, we convolve it with itself 4 times.")
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, sharey = True)
plt.xlim(0,20)
mu = 1
x = np.arange(5)
pdf = stats.rayleigh.pdf(x)
ax1.step(x,pdf)
pdf2 = signal.convolve(pdf, pdf)
x = np.arange(len(pdf2))
ax2.step(x,pdf2)
pdf3 = signal.convolve(pdf2, pdf2)
x = np.arange(len(pdf3))
ax3.step(x,pdf3)
pdf4 = signal.convolve(pdf3, pdf3)
x = np.arange(len(pdf4))
ax4.step(x,pdf4)
```

```python
print("After 4 convolution, we saw that the distribution is still a Rayleigh distribution, but the peak value is getting smaller.")
print(" ")
print("B. Discuss how the shape changes. Does it approach a Gaussian distribution? If yes, after how many intervals?")
print(" ")
print("After convolving for 4 times, the background distribution is not getting similar to Gaussian distribution, Thus, we would need to integrate it over a larger interval. Let's try 15 times.")
fig, ax = plt.subplots(1,1)
mu = 1
x = np.arange(5)
pdf = stats.rayleigh.pdf(x)
days = pdf
for x in range (15):
    days = signal.convolve(days, pdf)
    x = np.arange(len(days))
    ax.bar(x, days)
plt.show()
```

```python
print("Problem 4")
print(" ")
print("During previous question, we explored the statistical issue called a Trials Factor. Now, in this problem, we will do something different. Let us assume we have a Gaussian background and parameters just like in Problem 3, and we will invert the problem.")
print(" ")
print("A) Calculate the signal required for a 5-sigma detection in Version 1")
print(" ")
print("Given that our standard deviation in problem 3 is 5, which is the width of distribution, and the mean = 0, the signal we will get is equal to (mean + 5 * std), which is 25")
print(" ")
print("Let's check if we got the right signal strength. We first want to calculate the probability associated with 5 'sigma', which is ")
print(1 - stats.norm.cdf(25, scale = 5))
print(" ")
print("Then, we use ourr result to get the 'sigma' value, which is ")
print(stats.norm.ppf(1 - 2.866515719235352e-07, scale = 5))
print(" ")
print("It is consistant. Thus, a 5 'sigma' detection would need a signal strength of 25.")
print(" ")
```

```python
print("C)  Discuss how much brighter the signal must be for discoverey if you have a trials factor of 10k. Looking at your probability distributions, explain why the sensitivity penalty due to a trials factor is so low.")
print(" ")
r = stats.norm.ppf(1 - 2.866515719235352e-11, scale = 5) / stats.norm.ppf(1 - 2.866515719235352e-07, scale = 5)
print("If I have a trial factor of 10k, it requires the signal to be ", r, " , about 1.31, times brighter.")
print(" ")
print("Compared with the trial factor, we found that the sensitivity penalty is so much less than trial factor. The reason for this is that the sigma value is changing at a much lower ratio than the trial factor is.")
print(" ")
```

```python
print("D) If you changed the trials factor significantly (orders of magnitude), how large is the effect on your 5-sigma sensitivity threshold?")
print(" ")
print("Let us try a trail factor of 100k, and 10000k")
print(" ")
print("We first calculate the probability associated with 5 'sigma' for 100k trial factor")
print("prob = 2.866515719235352e-07 / 100000")
prob = 2.866515719235352e-07 / 100000
print("     =", prob)
print(" ")
print("Then, we use ourr result to get the 'sigma' value, which is ")
print(stats.norm.ppf(1 - 2.866515719235352e-12, scale = 5))
print(" ")
print("Thus, we will have a signal strength of 34.43, which is 1.37 times brighter.")
print(" ")
print("Now, let us try a trial factor of 10000k")
print("prob = 2.866515719235352e-07 / 10000000")
prob = 2.866515719235352e-07 / 10000000
print("     =", prob)
print(" ")
print("Then, we use ourr result to get the 'sigma' value, which is ")
print(stats.norm.ppf(1 - 2.866515719235352e-14, scale = 5))
print(" ")
print("Thus, we will have a signal strength of 37.57, which is 1.5 times brighter.")
print(" ")
print("From the results shown above, we can conclude that, if we change the trials factor significantly, their relationship is logarithmic. That is the sensitivity threshold is increasing with a raise of power of 10 of the trials factor.")
```
