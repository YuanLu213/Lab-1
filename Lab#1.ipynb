```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

#this sets the size of the plot to something useful
plt.rcParams["figure.figsize"] = (15,10)
```

```python
print("A Little Statistics")
print("1.")
print("B.")
s1 = stats.norm.cdf(1,loc = 0, scale = 1)
print(s1)
s2 = stats.norm.cdf(2, loc = 0, scale = 1)
print(s2)
s3 = stats.norm.cdf(3, loc = 0, scale = 1)
print(s3)
```

```python
print("As stated by the z-table, all our results are matched with the values it provided")
```

```python
print("C.")
print(stats.norm.ppf(0.8413447460685429, loc = 0, scale = 1))
print(stats.norm.ppf(0.9772498680518208, loc = 0, scale = 1))
print(stats.norm.ppf(0.9986501019683699, loc = 0, scale = 1))
print("As shown above, all the 'sigma' value we got for certain probability we took is the correct value.")
```

```python
print("2.")
print("A.")
d = stats.lognorm.rvs(0.9999, loc = 4.96, scale = 0.01, size = 10000)
print("B.")
fig, ax = plt.subplots(1, 1)
ax.hist(d,50, density=True)
plt.tick_params(labelsize = 24)
plt.xlim([4.95,5.05])
x = np.linspace(4.95,5.05,1000)
ax.plot(x,stats.lognorm.pdf(x, 0.9999, loc = 4.96, scale = 0.01),linewidth = 8,alpha = 0.7)
plt.show()
```

```python
print("3.")
print("B.")
print("I would like to know what is the probability to get a value of 4.99")
print(1 - stats.lognorm.cdf(4.99, 0.9999, loc = 4.96, scale = 0.01))
```

```python
print("E.")
print("What is the 'sigma' value for the probability above?")
print(stats.norm.ppf(0.13594463664421008, loc = 0,  scale = 1))
```

```python
print("4.")
print(1 - stats.lognorm.cdf(5, 0.9999, loc = 4.96, scale = 0.01))
print(stats.norm.ppf(0.0828073621504728, loc = 0,  scale = 1))
print(1 - stats.lognorm.cdf(5.02, 0.9999, loc = 4.96, scale = 0.01))
print(stats.norm.ppf(0.03657139591373737, loc = 0,  scale = 1))
print(1 - stats.lognorm.cdf(4.97, 0.9999, loc = 4.96, scale = 0.01))
print(stats.norm.ppf(0.5000000000000084, loc = 0,  scale = 1))
print("As the measurement is getting larger, the 'sigma' is still negative and becoming larger, and the probabilitlty will become smaller. As our measurements is getting smaller, the probability is getting alrger and the 'sigma' value becomes positive.")
```

```python
print("Non-continuous distributions")
print("1. A.")
fig, ax = plt.subplots(1, 1)

n, p = 10, 0.4
x = np.arange(stats.binom.ppf(0.01, n, p),
              stats.binom.ppf(0.99, n, p))
ax.plot(x, stats.binom.pmf(x, n, p), 'bo', ms=6, label='binom pmf')
ax.vlines(x, 0, stats.binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
```
