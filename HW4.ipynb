Problem 0: some fake data to mess with

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
plt.rcParams["figure.figsize"] = (20,15)
fig, ax = plt.subplots(1, 1)
ax.scatter(t, bkgd, color = "y")
ax.set_title("Scatter plot of all data", fontsize = 20)
ax.set_xlabel("Time", fontsize = 20)
ax.set_ylabel("Background", fontsize = 20)
```

Problem 1: subsets of data

1a) Make a scatter plot (not a line plot) of the first 1000 data points.

```python
fig, ax = plt.subplots(1, 1)

x = np.arange(0, 999)
arr = np.array(bkgd)
ax.scatter(x, arr[1:1000], color = "m")
ax.set_title("Scatter plot of first 1000 points", fontsize = 20)
ax.set_xlabel("Time", fontsize = 20)
ax.set_ylabel("Background", fontsize = 20)
```

1b) Generalize your code so you can make a plot of any X contiguous points (e.g. 2000 points starting at point 500), and produce an example plot of a set of data somethere in the middle of your array.

```python
X = 1500

fig, ax = plt.subplots(1, 1)
x = np.arange(1500, 1500 + X)
arr = np.array(bkgd)
ax.scatter(x, arr[1500:1500 + X], color = "b")
ax.set_title("Scatter plot of 1500 points starting at point 1500", fontsize = 20)
ax.set_xlabel("Time", fontsize = 20)
ax.set_ylabel("Background", fontsize = 20)
```

1c) Sometimes you want to sample the data, such as plotting every 100th point. Make a plot of the full data range, but only every 100th point.

```python
fig, ax = plt.subplots(1, 1)

x = np.arange(0, 100000, 100)
arr = np.array(bkgd)
ax.scatter(x, arr[1:100000:100], color = "c")
ax.set_title("Scatter plot of all points every 100th point", fontsize = 20)
ax.set_xlabel("Time", fontsize = 20)
ax.set_ylabel("Background", fontsize = 20)
```

Problem 2: density

2a) Make a 2d histogram plot. Read about options, and play with binsize so it is useful (too large a bin and everything washes out, too small and the sample noise dominates). Make sure you use a colorbar.

```python
fig, ax = plt.subplots(1, 1)

x = np.arange(0, 100000)
plt.hist2d(x, arr, bins = 70)
plt.colorbar()
plt.title("2D-histogram of all data", fontsize = 20)
plt.xlabel("Time", fontsize = 20)
plt.ylabel("Background", fontsize = 20)
```

2b) Clearly explain what is being plotted in your plot (e.g. what does each color mean?)
       
      Waht is being plotted is all 100000 data points in our sample data. The different color means the different data density. We can see that from the graph, the middle part of the graph is where data points are concentrated since, according to the color bar on the right, the more yellow the color gets, the higher the density is, and the darker and more blue the color gets, the lower the density is.
