Problem 1

For this problem, I generated a normal distributed background with a sigma value 3 and a uniform distributed signal with strength 15. 

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
plt.rcParams["figure.figsize"] = (15,10)

d = stats.norm.rvs(loc=0,scale=3, size = 1000000)
sig = stats.uniform.rvs(loc=0,scale=15, size = 1000000)

data = d + sig

signaledges = np.linspace(0,15,40)
dataedges = np.linspace(0,20,40)

Psd, temp, temp2= np.histogram2d(data,sig, bins=[dataedges,signaledges], density=True)

datacenters = (dataedges[:-1] + dataedges[1:]) / 2
signalcenters = (signaledges[:-1] + signaledges[1:]) / 2
plt.pcolormesh(datacenters,signalcenters,Psd.T)
plt.ylabel('True signal, $P(s|d)$', fontsize = 24)
plt.xlabel('Observed data, $P(d|s)$', fontsize = 24)
```

Atrue injected signal of 10 is injected to the data, and 𝑃(𝑑|𝑠) graph is plotted, which represents the probability of distribution of our data after a true signal of 10 is injected.

Problem 1c

An observed data value of 8 was observed and a graph of 𝑃(𝑠|𝑑) was plotted, which represents when we are reading a signal of 8 the probability distribution of our true signal injected. 

```python
obs_sig = 8
injected_sig = obs_sig - d
fx, ax = plt.subplots(1, 1)
ax.hist(injected_sig, 100, density=True, color = 'brown', histtype = 'step')
ax.set_title('P(signal|data)', fontsize = 16)
ax.set_xlabel('True Signal', fontsize = 16)
ax.set_ylabel('Probability', fontsize = 16)
```

Problem 2

Now we will repeat the above, but with a background with non-zero mean of 6.

```python
d = stats.norm.rvs(loc=6,scale=3, size = 1000000)
sig = stats.uniform.rvs(loc=0,scale=15, size = 1000000)

data = d + sig

signaledges = np.linspace(0,15,40)
dataedges = np.linspace(0,20,40)

Psd, temp, temp2= np.histogram2d(data,sig, bins=[dataedges,signaledges], density=True)

datacenters = (dataedges[:-1] + dataedges[1:]) / 2
signalcenters = (signaledges[:-1] + signaledges[1:]) / 2
plt.pcolormesh(datacenters,signalcenters,Psd.T)
plt.ylabel('True signal, $P(s|d)$', fontsize = 24)
plt.xlabel('Observed data, $P(d|s)$', fontsize = 24)
```

From the above 2D histogram, we can see that the center of the histogram is shifted towards right from 0 to approximately 6.

```python
fx, (ax1, ax2) = plt.subplots(1, 2)
newSignal2 = d + signaledges[26]
ax1.hist(newSignal2, 100, density=True, color = 'k', histtype = 'step')
ax1.set_title('P(data|signal)', fontsize = 16)
ax1.set_xlabel('Observed Signal', fontsize = 16)
ax1.set_ylabel('Probability', fontsize = 16)

obs_sig = 8
injected_sig = obs_sig - d
ax2.hist(injected_sig, 100, density=True, color = 'brown', histtype = 'step')
ax2.set_title('P(signal|data)', fontsize = 16)
ax2.set_xlabel('True Signal', fontsize = 16)
```

From the above separate graphs of 𝑃(𝑑|𝑠) and 𝑃(𝑠|𝑑), we can see that they are centered at different value of signal than in problem 1.
