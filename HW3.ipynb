2a) What is the pdf of the sum of two identical exponential distributions?

The pdf of sum of two identical exponential distributions is equal to λXe^-λx, which would be an Erlang distribution.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
from scipy import signal
#this sets the size of the plot to something useful
plt.rcParams["figure.figsize"] = (20,15)

fig,  ax = plt.subplots(1, 1)
plt.xlim(0,10)
x = np.linspace(0,50,500)
expo = stats.expon.pdf(x)
x2 = np.linspace(0,100,999)
conv = signal.convolve(expo,expo)
ax.plot(x,expo)
ax.plot(x2,conv)
plt.title("Probability Density Function")
plt.xlabel("X")
plt.ylabel("Probability")
plt.show()
```
